```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.array([0,1,2,3,4])
y=x**2
plt.plot(x,y)
print("=Gráfica sencilla")
```

```python
x=np.linspace(0,10,100)
y=x**2
plt.plot(x,y)
print'grafica sencilla'
```

```python
x=np.linspace(0,10,100)
y1=x
y2=x**2
plt.plot(x,y1)
plt.plot(x,y2)
print'==dos graficas sencillas'
```

```python
x=np.linspace(0,10,100)
y1=x
y2=x**2
plt.plot(x,y1,label='linea')
plt.plot(x,y2,label="cuadrado")
plt.legend()
print'==dos graficas sencillas=='
```

```python
x=np.linspace(0,10,100)
y1=x
y2=x**2
y3=np.sqrt(x)
y4=np.power(x,1.5)

plt.plot(x,y1,label='linea',linestyle='-')
plt.plot(x,y2,label='cuadrado',linestyle=':')
plt.plot(x,y3,label='Raiz',linestyle='-.')
plt.plot(x,y4,label='Potencia 1.5',linestyle='--')
plt.legend()
print'==dos graficas sencillas'
```

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
plt.scatter(x,y)
print'==scatter de puntos aleatoreos=='
```

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colores=np.random.rand(N)
radios=15*np.random.rand(N)
areas=np.pi*radios**2
plt.scatter(x,y, s=areas, c=colores, alpha=0.5)
print("== Scatter plot de puntos aleatorios ==")
```

```python
N=500
x=np.random.rand(N)
plt.hist(x)
print'Histograma aleatoreo'
```

```python
N=500
x=np.random.randn(N)
plt.hist(x)
print("== Histograma aleatorio Normal ==")
```

```python
N=1000
x1=np.random.randn(N)
x2=2+2*np.random.randn(N)

plt.hist(x1,20, alpha=0.3)
plt.hist(x2,20, alpha=0.3)
print("== Histograma de dos distribuciones ==")
```

```python
N=10
matriz = np.random.rand(N,N)
plt.imshow(matriz)
print'=Imagen de matriz='
```

```python
N=10
matriz = np.random.rand(N,N)
plt.imshow(matriz, interpolation='nearest')
plt.grid()
print'=Imagen de matriz='
```
